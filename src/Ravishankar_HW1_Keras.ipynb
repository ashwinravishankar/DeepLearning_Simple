```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.optimizers import SGD
from keras import losses
from keras import backend as K
from csv import reader
import numpy as np
import matplotlib.pyplot as plt
```

```python
def readFile(fileName):
    trainData = []
    with open(fileName, 'r') as file:
        csvReader = reader(file)
        next(csvReader, None)
        for row in csvReader:
            for i in range(1, 3):
                row[i] = float(row[i].strip())
            row[0] = int(row[0].strip())
            trainData.append(row)
    x_data = []
    y_data = []
    for i in range(len(trainData)):
        temp_arr = []
        x_data.append(trainData[i][1:])
        if (trainData[i][0] == 0):
            temp_arr = [1, 0]
        else:
            temp_arr = [0, 1]
        y_data.append(temp_arr)
    return np.asarray(x_data), np.asarray(y_data)

x_train, y_train = readFile ("train.csv")
x_test, y_test = readFile("test.csv")
```

```python
def squared_loss_function(y_true, y_pred):
    return K.square(y_pred - y_true)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
model = Sequential()

model.add(Dense(3, input_dim=2, use_bias=False, activation='sigmoid'))
model.add(Dense(3, use_bias=False, activation='sigmoid'))
model.add(Dense(2, use_bias=False, activation='softmax'))
```

```python
model.compile(loss=squared_loss_function, optimizer='sgd', metrics=['accuracy'])
```

```python
model_metrics = model.fit(x_train, y_train, epochs=300, batch_size=None)
```

```python
score = model.evaluate(x_test, y_test)
print(score)
print("\n\nLearned Weights\n")
for layer in model.layers:
#     print(layer.get_config(), layer.get_weights())
    print(layer.get_weights())
    print("\n")
```

```python
print(model_metrics.history.keys())
```

```python
plt.plot(model_metrics.history['acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train'], loc='lower right')
plt.show()
```

```python
plt.plot(model_metrics.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper right')
plt.show()
```

```python
pred = model.predict(x_test, batch_size=None)
truePositive=0
trueNegative=0
falsePositive=0
falseNegative=0
# print(y_test[0][1])
for i in range(len(x_test)):
#     print(i+1, pred[i], 0 if (pred[i,0] > pred[i,1]) else 1, )
    if(y_test[i][0]==1):
        if (pred[i,0]>pred[i,1]):
            trueNegative +=1
        else:
            falseNegative+=1
    elif(y_test[i][1]==1):
        if (pred[i,0]<pred[i,1]):
            truePositive +=1
        else:
            falsePositive+=1
accuracy = ((truePositive + trueNegative) / len(x_test))*100
precision=(truePositive)/(truePositive+falsePositive)
recall=(truePositive)/(truePositive+falseNegative)
f1score=2*((precision*recall)/(precision+recall))
print("Test Metrics:")
print("   Total in test set:", len(x_test))
print("   Correctly predicted:", truePositive+trueNegative)
print("   Accuracy: ", accuracy)
print("   Precision: ", precision)
print("   Recall: ", recall)
print("   F1-Score: ", f1score)
```
